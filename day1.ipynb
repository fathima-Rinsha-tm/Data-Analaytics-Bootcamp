```python
import csv

students = [
    ["Name", "Age", "Grade"],
    ["rinu", 20, "A"],
    ["sugu", 21, "B"]
]

# Write to CSV
with open("students.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(students)
with open("students.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)


with open("students.csv","a",newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["joseph",20,"A"])
    writer.writerow(["mary",22,"B"])
with open("students.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
         print(row)




```

```python
# Create a text file
file = open("textfile.txt", "w")
file.write("Alice is a student. Alice likes Python. Alice studies well.")
file.close()

# Open the file in read mode
file = open("textfile.txt", "r")

# Read the file content
data = file.read()

# Count occurrences of the word "Alice"
count = data.count("Alice")

# Display the result
print("Number of occurrences of Alice:", count)

# Close the file
file.close()
```

```python
# Create a text file
file = open("textfile.txt", "w")
file.write("The Queen is kind.\n")
file.write("Alice is a student.\n")
file.write("The Queen lives in a palace.\n")
file.write("Python is easy.\n")
file.close()

# Open the file in read mode
file = open("textfile.txt", "r")

# Initialize count
count = 0

# Read file line by line
for line in file:
    if "Queen" in line:
        count = count + 1

# Display result
print("Number of lines containing 'Queen':", count)

# Close the file
file.close()
```

```python
# Create a text file
file = open("textfile.txt", "w")
file.write("Alice likes Python Alice likes coding")
file.close()

# Open the file in read mode
file = open("textfile.txt", "r")

# Read the content
data = file.read()

# Split into words
words = data.split()

# Find unique words
unique_words = set(words)

# Display each unique word and its count
for word in unique_words:
    print(word, ":", words.count(word))

# Close the file
file.close()
```

```python
# Open the book file in read mode
file = open("book.txt", "r")

# Read the contents
data = file.read()

# Split the content into words
words = data.split()

# Assume the first word is the longest
longest_word = words[0]

# Check each word
for word in words:
    if len(word) > len(longest_word):
        longest_word = word

# Display the longest word
print("Longest word:", longest_word)
print("Length:", len(longest_word))

# Close the file
file.close()
```
